```python
import cv2
from google.colab.patches import cv2_imshow
```

**Change Image's colour profile**

```python
img = cv2.imread("images/color.jpg")
```

```python
cv2_imshow(img)
```

```python
img.shape
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray.shape, cv2_imshow(gray)
```

```python
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
hsv.shape, cv2_imshow(hsv)
```

**Edge Detection**

```python
edge = cv2.Canny(gray, 150, 200)
cv2_imshow(edge)
```

```python
import numpy as np
```

```python
#erosion and dilation
kernel = np.ones((5,5), np.uint8)
erode_image = cv2.erode(edge, kernel, iterations = 1)
cv2_imshow(erode_image)
```

```python
dilation = cv2.dilate(edge, kernel, iterations = 1)
cv2_imshow(dilation)
```

```python
erode_on_dilation_image = cv2.erode(dilation, kernel, iterations = 1)
cv2_imshow(erode_on_dilation_image)
```

```python
display = np.hstack((edge, dilation, erode_on_dilation_image))
cv2_imshow(display)
```
