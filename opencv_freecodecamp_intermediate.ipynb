Colour Detection

```python
img = cv2.imread("images/shapes.png")
img.shape
```

```python
#BGR Image
cv2_imshow(img)
```

```python
hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
hsv_img.shape
```

```python
display = np.hstack((img, hsv_img))
cv2_imshow(display)
```

```python
#check trackbar in opencv for upper & lower bound
#blue colour triangle
lower_blue = np.array([65,0, 0])
upper_blue = np.array([110,255,255])

#red colour circle
lower_red = np.array([0 ,0, 0])
upper_red = np.array([20,255,255])

#green colour
lower_green = np.array([46 ,0, 0])
upper_green = np.array([91,255,255])

#yellow colour
lower_yellow = np.array([21 ,0, 0])
upper_yellow= np.array([45,255,255])

#red & green colour
lower_red_green = np.array([0 ,0, 0])
upper_red_green = np.array([91,255,255])

mask1 = cv2.inRange(hsv_img, lower_blue, upper_blue)
mask2 = cv2.inRange(hsv_img, lower_red, upper_red)
mask3 = cv2.inRange(hsv_img, lower_green, upper_green)
mask4 = cv2.inRange(hsv_img, lower_red_green, upper_red_green)
mask5 = cv2.inRange(hsv_img, lower_yellow, upper_yellow)


#cv2_imshow(mask1)
#cv2_imshow(mask2)
#cv2_imshow(mask3)
#cv2_imshow(mask4)
cv2_imshow(mask5)
```

```python
result1 = cv2.bitwise_and(img, img, mask = mask1)
result2 = cv2.bitwise_and(img, img, mask = mask2)
result3 = cv2.bitwise_and(img, img, mask = mask3)
result4 = cv2.bitwise_and(img, img, mask = mask4)
result5 = cv2.bitwise_and(img, img, mask = mask5)


#cv2_imshow(result1)
#cv2_imshow(result2)
#cv2_imshow(result3)
#cv2_imshow(result4)
cv2_imshow(result5)
```
